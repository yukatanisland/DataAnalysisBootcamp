```python
bins = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6]

# Create names for the four groups
group_names = ['0', '0.5','1', '1.5','2','2.5', '3','3.5', '4','4.5','5','5.5']

#make new column with letter grades
df['Bincarat'] = pd.cut(df['carat'], bins, labels=group_names)
df.head()
```

```python
df.groupby(['Bincarat','color']).mean()
#color and price have a correlation.
```

```python
def score_to_numeric(x):
    if x=='IF':
        return 1
    if x=='VVS1':
        return 2
    if x=='VVS2':
        return 3
    if x=='VS1':
        return 4
    if x=='VS2':
        return 5
    if x=='SI1':
        return 6
    if x=='SI2':
        return 7
    if x=='I1':
        return 8     
    
df['clarity_val'] = df['clarity'].apply(score_to_numeric)
df.groupby(['Bincarat','clarity_val']).mean()
```

```python
df.groupby(['Bincarat','cut']).mean()
```

```python
def score_to_numeric(x):
    if x=='Ideal':
        return 1
    if x=='Premium':
        return 2
    if x=='Very Good':
        return 3
    if x=='Good':
        return 4
    if x=='Fair':
        return 5
    
df['cut_val'] = df['cut'].apply(score_to_numeric)
df.groupby(['Bincarat','cut_val']).mean()
#cut and price have a correlation.
```
